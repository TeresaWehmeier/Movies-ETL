```python
def clean_movie(movie):
    movie = dict(movie) #create a non-destructive copy
    alt_titles = {}
    # combine alternate titles into one list
    for key in ['Also known as','Arabic','Cantonese','Chinese','French',
                'Hangul','Hebrew','Hepburn','Japanese','Literally',
                'Mandarin','McCune-Reischauer','Original title','Polish',
                'Revised Romanization','Romanized','Russian',
                'Simplified','Traditional','Yiddish']:
        if key in movie:
            alt_titles[key] = movie[key]
            movie.pop(key)
    if len(alt_titles) > 0:
        movie['alt_titles'] = alt_titles

    # merge column names
    def change_column_name(old_name, new_name):
        if old_name in movie:
            movie[new_name] = movie.pop(old_name)
    change_column_name('Adaptation by', 'Writer(s)')
    change_column_name('Country of origin', 'Country')
    change_column_name('Directed by', 'Director')
    change_column_name('Distributed by', 'Distributor')
    change_column_name('Edited by', 'Editor(s)')
    change_column_name('Length', 'Running time')
    change_column_name('Original release', 'Release date')
    change_column_name('Music by', 'Composer(s)')
    change_column_name('Produced by', 'Producer(s)')
    change_column_name('Producer', 'Producer(s)')
    change_column_name('Productioncompanies ', 'Production company(s)')
    change_column_name('Productioncompany ', 'Production company(s)')
    change_column_name('Released', 'Release Date')
    change_column_name('Release Date', 'Release date')
    change_column_name('Screen story by', 'Writer(s)')
    change_column_name('Screenplay by', 'Writer(s)')
    change_column_name('Story by', 'Writer(s)')
    change_column_name('Theme music composer', 'Composer(s)')
    change_column_name('Written by', 'Writer(s)')

    return movie
```

```python
# add additional patterns in the box office found in above results not found in either matches.
# spaces after dollar sign = \s*
# use periods as thousands separator instead of comma - form 2 add (?:[,\.]\d{3}+ to replace (?:,\d{3})+
    # will also match on 1.234 billion, but only trying to find $123.456.789. Don't want to capture 1.234 billion
    # so add negavtive lookahead group at end of form 2

form_one = r'\$\s*\d+\.?\d*\s*[mb]illi?on'
form_two = r'\$\s*\d{1,3}(?:[,\.]\d{3})+(?!\s[mb]illion)'
```

```python
# create a function to turn the extracted values into a numeric value using parse_dollars; take a string and return
    #a floating-point number. 
    
def parse_dollars(s):
    # if s is not a string, return NaN
    if type(s) != str:
        return np.nan
    
    # if input is of the form $###.# million
    if re.match(r'\$\s*\d+\.?\d*\s*milli?on', s, flags=re.IGNORECASE):
        
        
        #remove dollar sign and ' million'
        s= re.sub('\$|\s|[a-zA-Z]','',s)
        
        # convert to float and mltply by a million
        value = float(s) * 10**6
        
        # return value
        return value
        
    # if input is of the form $###.# billion
    elif re.match(r'\$\s*\d+\.?\d*\s*billi?on', s, flags=re.IGNORECASE):
        
        
        # remove dollar sign and ' billion'
        s= re.sub('\$|\s|[a-zA-Z]','',s)

        # convert to float and multiply by a billion
        value = float(s) * 10**9

        # return value
        return value
    
    # if input is of the form $###,###,### 
    elif re.match(r'\$\s*\d{1,3}(?:[,\.]\d{3})+(?!\s[mb]illion)', s, flags=re.IGNORECASE):
    
        # remove dollar sign and commmas
        s= re.sub('\$|,','',s)
        
        # convert to a float
        value = float(s)
        
        # return value
        return value
        
    # otherwise, return NaN
    
    else:
        return np.nan
```

```python
#regex character types:
# \d - matches any digit from 0 to 9
# \D matches any non-digit character
# \w matches a word character (a letter, digit, or underscore)
# \W matches any non-word character (anthing other than a letter, or underscore such as spaces and punctuation)
# \s will match any whitespace character (including spaces, tabs and newlines)
# \S will match any non-whitespace characters
# [] represents series: can include a character type (above) inside the series; find million or billion
# by using series '[mb]illion'.
# period (.) is a wildcard and matches any single character alpha, numberic whitespace or punctuation
# Escaping \ - tells parser to treat upcoming metacharacters like a literal character. "ca\." matches only "ca."
# Special counting characters (prior to this character): *, + {},?
# * previous character can repeat any number of times including zero 'ca*t" = 'cat','caaat','fiction' can be returned
# + means previous has to show at least once, so 'ca*t', 'caaat', but not fiction
# {} means it will show up an exact number of times so "\d{7}" means seven digits must occur; "ca{3}t"
#    means only 'caaat' will show because it must occur three times; or use range "ca{3,5}t" matches 'caaat','caaaaat', not cat
# ? is for optional characters, so it can show zero or once. "ca?t" is = to "ca{0,1}t".
# Alternation: | (pipe) is logical OR. cat or mouse or dog = 'cat|mouse|dog'.
# String boundaries = ^ and $. ^ means only at the beginning of a string; $ means only at the end of a string.
# '^cat' would mathc 'cat' and 'catatonic', but not 'concatenate'. 'cat$' would match 'cat' and 'tomcat' not catatonic
# string boundaries represent zero length matches; don't match any actual characters tehmselves, just the boundaries of string
        # searched.
# Capture groups (). Adds structure to a search pattern; '1,000','1,000,000','1,000,000,000' as strings are similar
    # regular expression that matches all three is '1(,000)+'
    # "capture groups" define WHAT info should be extracted. Example: '\d{3}-\d{3}-\d{4}' looks for a phone no. but
    # if we just want to capture the digits we use '(\d{3})-(\d{3})-(\d{4})' and only the phone digits are returned.
# Non-capturing groups ?:
    # Example, want to change prefix in phone # to a different one. Using non-capture '(?:\d{3})-(\d{3})-(?:\d{4})'
    # will only return the prefix values after that don't include ?:. It searches for the pattern but ignores anything
    # with ?: inside parethenses. If area code were to be missing, it would not catch prefix, since it doesn't fit the pattern.
# Negative look ahead groups: example phone 333-333-4444, but an id uses patter 333-333-55555. '(\d{3})-(\d{3})-(\d{4})' will
    # think it is a phone no and just ignore the last digit. But use ?! in the parentheses and it will exclude anthing with a
    # single digit immediately following the patter ((\d{3})-(\d{3})-(\d{4}(?!\d)))
```

```python
# create a function that fills in missing data for a column pair then drops the redundant column

def fill_missing_kaggle_data(df, kaggle_column, wiki_column):
    df[kaggle_column] = df.apply(
        lambda row: row[wiki_column] if row [kaggle_column] == 0 else row[kaggle_column]
        , axis=1)
    df.drop(columns=wiki_column, inplace=True)
```

```python
fill_missing_kaggle_data(movies_df, 'runtime','running_time')
fill_missing_kaggle_data(movies_df, 'budget_kaggle','budget_wiki')
fill_missing_kaggle_data(movies_df, 'revenue','box_office')
movies_df
```
